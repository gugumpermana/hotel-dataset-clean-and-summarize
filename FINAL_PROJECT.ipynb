## 8. Visualize the arrival_date_month to see in what month the most visitors come


```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
arrival_months_group = data3.groupby("arrival_date_month")[["adults","children","babies"]].agg(["sum"])
arrival_months_group["total"] = arrival_months_group.sum(axis=1)
months_sort_list = ['January','February','March','April','May','June','July','August','September','October','November','December']
arrival_months_group.index = pd.Categorical(arrival_months_group.index, categories=months_sort_list, ordered=True)
arrival_months_group = arrival_months_group.sort_index()
month_list = arrival_months_group.index.tolist()
total_visitor_list = arrival_months_group["total"].tolist()

month_list.reverse()

plt.barh(month_list, total_visitor_list)

plt.title("Visitor Attendant each Month")
#plt.xlabel ("Month")
plt.xlabel("Number of Visitors")
plt.tight_layout()
plt.show()
```

### Conclusion: the most visited month is in August, more than 30000 Visitors
